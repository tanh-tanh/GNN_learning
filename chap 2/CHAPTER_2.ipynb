# Introducing graph properties

in networkx we can define undirected graph with nx.Graph and directed graph with nx.DiGraph 

````python
import networkx as nx
import matplotlib.pyplot as plt

g = nx.Graph()
g.add_edges_from([('A','B'),('B','C'),('C','D'),('D','B')])

plt.axis('off')
nx.draw_networkx(g)
````

````python
dg = nx.DiGraph()
dg.add_edges_from([('A','B'),('B','D'), ('C','B'), ('C','A')])

nx.draw_networkx(dg)
````

## weighted graph:
we can store like undirect or direct but add weight: ('A','B',{'weight: 100'})

````python
WG = nx.Graph()
WG.add_edges_from([('A', 'B', {"weight": 10}), ('A', 'C', {"weight": 20}), ('B', 'D', {"weight": 30}), ('B', 'E', {"weight": 40}), ('C', 'F', {"weight": 50}), ('C', 'G', {"weight": 60})])
labels = nx.get_edge_attributes(WG, "weight")

plt.axis('off')
nx.draw_networkx(WG,
                 pos=nx.spring_layout(WG, seed=0),
                 )
nx.draw_networkx_edge_labels(WG, pos=nx.spring_layout(WG, seed=0), edge_labels=labels)
````

# Connected graph

Connected graph every pair of u and v there are always have a path form u to v
check with is_connected()

````python
g = nx.Graph()
g.add_edges_from([('A','B'), ('B','C'), ('D','E')])

print(f"is g is a connect graph {nx.is_connected(g)}")
nx.draw_networkx(g)
````

````python
g = nx.Graph()
g.add_edges_from([('A','B'), ('B','C'), ('D','A')])

print(f"is g is a connect graph {nx.is_connected(g)}")
nx.draw_networkx(g)
````

# graph concept:
## fundermental obj:
Degree of node (deg(node)): number of edge incident to this node\
In undirected : number of edge connect to it\
In directed: indgree, outdgree (ifykyk)\
We can use nx.degree for (undirected), and nx.in_degree, nx.out_degree

````python
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
print(f"deg(A) = {G.degree['A']}")

DG = nx.DiGraph()
DG.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')])
print(f"deg^-(A) = {DG.in_degree['A']}")
print(f"deg^+(A) = {DG.out_degree['A']}")
````

## graph meansures

Degree centrality: is just degree of node?. a high degree mean that node highly connected to other node, and significantly influences the network\
Closeness centrality: how close the node is to other node. It corresponds to avg  lenght of shortest path between all taget node and all other node. If it high it mean that node can quickly reach other node\
Between centrality: numbertime of node that lie in shortest path between pair of other node. If it high it mean that's bottle neck\


````python
print(f"Degree centrality      = {nx.degree_centrality(G)}")
print(f"Closeness centrality   = {nx.closeness_centrality(G)}")
print(f"Betweenness centrality = {nx.betweenness_centrality(G)}")
````

## density
```
ratio between acutal number of edge and maximum number of edge.
```
A graph with high density is considered more information flow than low\
For an undirected graph maximum number of edge is $\frac{n*(n-1)}{2}$\
For a directed graph maximum numbder of edge is $n*(n-1)$ \
Generally, a graph is dense if density greater than 0.5 and spare if its density less then 0.1
